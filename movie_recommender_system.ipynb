### Data Pre processing

```python
import ast
ast.literal_eval('[{"id": 1463, "name": "culture clash"}, {"id": 2964, "name": "future"}, {"id": 3386, "name": "space war"}, {"id": 3388, "name": "space colony"}, {"id": 3679, "name": "society"}, {"id": 3801, "name": "space travel"}, {"id": 9685, "name": "futuristic"}, {"id": 9840, "name": "romance"}, {"id": 9882, "name": "space"}, {"id": 9951, "name": "alien"}, {"id": 10148, "name": "tribe"}, {"id": 10158, "name": "alien planet"}, {"id": 10987, "name": "cgi"}, {"id": 11399, "name": "marine"}, {"id": 13065, "name": "soldier"}, {"id": 14643, "name": "battle"}, {"id": 14720, "name": "love affair"}, {"id": 165431, "name": "anti war"}, {"id": 193554, "name": "power relations"}, {"id": 206690, "name": "mind and soul"}, {"id": 209714, "name": "3d"}]'
)###converting the string of list into list of dictionaries to fetch the names
```

```python
movies['keywords'].values[0]
```

```python
###Function to fetch out names from the dictionary
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
####function for cast column
def convert3(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            l.append(i['name'])
            counter+=1
        else:
            break
    return l
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
##function to fetch director name from crew column
def fetch_director(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
movies['crew']
```

```python
###converting string of overview column into list of strings by splitting every word
movies['overview']=movies['overview'].apply(lambda x: x.split())
```

```python
movies.head()
```

```python
#### Now we have to delete the space between words to avoid confusion to create tags. This will help in creating individual 
##unique tags
```

```python
movies['genres']=movies['genres'].apply(lambda x: [i.replace(' ','') for i in x])
movies['keywords']=movies['keywords'].apply(lambda x: [i.replace(' ','') for i in x])
movies['crew']=movies['crew'].apply(lambda x: [i.replace(' ','') for i in x])
movies['cast']=movies['cast'].apply(lambda x: [i.replace(' ','') for i in x])
```

```python
movies.head()
```

```python
###concatenating  genres, overview, cast, crew column to create tags column
movies['tags']=movies['genres'] + movies['overview'] + movies['cast'] + movies['crew'] 
movies['tags'][0]
```

```python
movies.head()
```

```python
new_df=movies[['id','title','tags']]  ##new required dataframe 
new_df
```

```python
###  we have to convert the list of strings in the tags column to string
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
new_df.head()
```

```python
new_df['tags'][0]
```

```python
##we should convert the tags into lower case
new_df['tags']=new_df['tags'].apply(lambda x: x.lower())
new_df['tags'][0]
```

##### Vectorization for calculating similarity between different movies

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
vectors[0]
```

```python
vectors.shape###4806 novies 5000 mostly repeated words
```

```python
### we can get the names of all the 500 words
cv.get_feature_names()
```

##### Now vectors of all movies are created.

##### Now we have to find the distance of all movies with each other. More distance between movies means less similarity between them

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
similarity
```

```python
cosine_similarity(vectors).shape
```

```python
similarity[0] ##similarity score of book with index value 0
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key= lambda x: x[1])[1:6] ##(index,distance)
```

```python
def recommend(movie):
    #fetch index
    movie_index=new_df[new_df['title']==movie].index[0]
    distance=similarity[movie_index]
    movie_list=sorted(list(enumerate(distance)),reverse=True,key= lambda x: x[1])[1:6]
    
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('Batman Begins')
```

```python
recommend("Pirates of the Caribbean: At World's End")
```
