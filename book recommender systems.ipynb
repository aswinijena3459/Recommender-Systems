### Popularity based recommender system

```python

ratings_with_name=ratings.merge(books,on='ISBN')
```

```python
ratings_with_name
```

#### We will take average rating of top 50 books.Books with minimum 250 no.of votes will be taken

#### creating new dataframe to calculate no of ratings on each book

```python
num_ratings_df=ratings_with_name.groupby(['Book-Title']).count()['Book-Rating'].reset_index()
```

```python
num_ratings_df.rename(columns={'Book-Rating':'num_of_ratings'},inplace=True)
num_ratings_df
```

```python
#### Cretaing new data frame for calculating average rating of every book
avg_rating_df=ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

#### Now merge both dataframes on book tiutl column

```python
popularity_df=num_ratings_df.merge(avg_rating_df,on='Book-Title')
popularity_df
```

#### Now filtering out books with more than 2500 no of ratings and selecting top 50 highly rated books

```python
top_50_books=popularity_df[popularity_df['num_of_ratings']>=250].sort_values('avg_rating',ascending=False).head(50)
top_50_books
```

#### I want to show publisher name ,book title ,image,author of book in my list of popular 50 books

```python
#### merging and dropping duplicates because there are diferent isbn numbers of same book
top_50_books=top_50_books.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_of_ratings','avg_rating']]
top_50_books
```

## Collaberative Filter based Recommender system

#### Here we have to create a new dataset where i can analyse ratings of all different users on different books each

#### We will take books which have got minimum 50 no of ratings and users who have given minimum 200 no of ratings, just to increase the efficiency of this recommender system

```python
x=ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
x
```

```python
### got the index/user ids of users whohave given more than 200 ratings
best_users=x[x].index
best_users
```

```python
###dataframe filtered on the basis of users who have given more than 200 ratings 
filtered_ratings=ratings_with_name[ratings_with_name['User-ID'].isin(best_users)]
filtered_ratings
```

```python
### Now we will filter out famous books who have more than 50 ratings on it
y=filtered_ratings.groupby('Book-Title').count()['Book-Rating'] >=50
famous_books=y[y].index
```

```python
famous_books
```

```python
final_ratings=filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
final_ratings
```

```python
final_ratings.drop_duplicates()
```

```python
###creating the required dataframe using pivot table and filling the nan values with 0
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt.fillna(0,inplace=True)
```

```python
pt
```

#### Now we will try to calculate the euclidian didtance of all popular books with each other to find the similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pt)
similarity_score
```

```python
##shape of similarity scores
similarity_score.shape
```

```python
similarity_score[0]###similarity score book named 1984 with other 705 books
```

#### Now we will try to create a function which will give us top 5 simillar books

```python
def recommend(book_name):
    #index fetch
    index=np.where(pt.index==book_name)[0][0]
    #top5 books according to closest distance from given book
    simillar_books=sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    
    print('You can also read:-')
    for i in simillar_books:
        print(pt.index[i[0]])
    
```

```python
recommend('1st to Die: A Novel')
```

```python
recommend('A Bend in the Road')
```

```python
recommend('You Belong To Me')
```
